```python
def corrfunc(x,y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    r, _ = pearsonr(x, y)
    ax = ax or plt.gca()
    # Unicode for lowercase rho (ρ)
    rho = '\u03C1'
    ax.annotate(f'{rho} = {r:.2f}', xy=(.1, .9), xycoords=ax.transAxes)
    
colors = iter(['xkcd:aqua', 'xkcd:coral', 'xkcd:purple blue',
       'xkcd:lightgreen', 'xkcd:orchid',
       'xkcd:yellow', 'xkcd:dusky purple', 'xkcd:violet',
       'xkcd:bright sky blue', 'xkcd:green', 'xkcd:brownish',
       'xkcd:moss green', 'xkcd:deep blue', 'xkcd:melon',
       'xkcd:faded green', 'xkcd:cyan', 'xkcd:brown green',
       'xkcd:purple blue', 'xkcd:green', 'xkcd:greyish blue'])

def my_scatter(x,y, **kwargs):
    kwargs['color'] = next(colors)
    plt.scatter(x,y, **kwargs)

def my_hist(x, **kwargs):
    kwargs['color'] = next(colors)
    plt.hist(x, **kwargs)

g = sns.pairplot(new_df, corner=True, kind="reg")
g.map_lower(corrfunc)
g.map_offdiag(my_scatter)


plt.savefig("Nutrient_correlation.jpg", bbox_inches='tight')
```
